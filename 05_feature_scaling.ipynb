```python
import seaborn as sns
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
#feature scaling
#Standardisation >> mu=0 and sigma 1
#Normalization (minmax scaling) >> data comes between 0 and 1
#unit vector
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
mean = np.mean(df['total_bill'])
```

```python
std = np.std(df['total_bill'])
```

```python
std_data = []
for i in list(df['total_bill']):
    zscore = (i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

```python
sns.distplot(df[['total_bill']])
```

```python
sns.distplot(std_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler =  StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']]) #it will calculate mu and sigma
```

```python
scaler.transform(df[['total_bill']])
```

```python
df
```

```python
scaler.fit_transform(df[['total_bill']])
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill', 'tip']]), columns = ['total_bill', 'tip'])
```

```python
scaler.transform([[13,1]])
```

```python
#Normalization/minmax scaling

from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max.fit_transform(df[['total_bill','tip']])
```

```python
min_max.transform([[10, 2]])
```

```python
#unit vector

from sklearn.preprocessing import normalize
```

```python
uv = normalize(df[['total_bill']])
```
