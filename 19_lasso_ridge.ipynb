```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = sns.load_dataset('mpg')
```

```python
df
```

```python
df.drop('name',axis = 1,inplace = True)
```

```python
df.isnull().sum()
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
df['origin'].value_counts()
```

```python
df['origin'] = df['origin'].map({'usa':1,'japan':2,'europe':3})
```

```python
df['origin'] = df['origin'].astype(int)
```

```python
df.dtypes
```

```python
#seprate x and y
x = df.drop('mpg', axis =1)
y = df['mpg']
```

```python
x
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1,random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
#simple linear regression model
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
```

```python
regression_model
```

```python
regression_model.fit(x_train,y_train)
```

```python
for i,col_name in enumerate(x_train.columns):
    print(f'the coff for {col_name} is {regression_model.coef_[i]}')
```

```python
#coefficients are relatively smaller, if one independent variable changes 
#slightly there will be not much difference in prediction.
#This is sometime is called smoother model
```

```python
from sklearn.metrics import r2_score

y_pred_linear = regression_model.predict(x_test)
r2_linear = r2_score(y_test, y_pred_linear)
print(f"R square of linear regression {r2_linear}")
```

```python
#ridge regression
from sklearn.linear_model import Ridge
ridge_regression_model = Ridge(alpha = 0.1)
ridge_regression_model.fit(x_train,y_train)

for i,col_name in enumerate(x_train.columns):
    print(f'the coff for {col_name} is {ridge_regression_model.coef_[i]}')
```

```python
# For Ridge Regression evaluation
y_pred_reg = ridge_regression_model.predict(x_test)
r2_ridge  = r2_score(y_test,y_pred_reg)
print(f"R-squared score for Ridge Regression: {r2_ridge}")
```

```python
#lasso regression
from sklearn.linear_model import Lasso
lasso_redge_model = Lasso(alpha = 0.5)
lasso_redge_model.fit(x_train,y_train)

for i, col_name in enumerate(x_train.columns):
    print(f"The coefficient for {col_name} is {lasso_redge_model.coef_[i]}")
```

```python
#3 features coefficients are 0, Lasso helps in feature selection
```
